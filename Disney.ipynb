````python
import pandas as pd


seven_dwarfs_df = pd.read_csv("./Disney Resources/7_dwarfs_train.csv")
alien_saucers_df = pd.read_csv("./Disney Resources/alien_saucers.csv")
entities_df = pd.read_csv("./Disney Resources/entities.csv")
metadata_df = pd.read_csv("./Disney Resources/metadata (1).csv")
pirates_of_caribbean_df = pd.read_csv("./Disney Resources/pirates_of_caribbean.csv")
rock_n_rollercoaster_df = pd.read_csv("./Disney Resources/rock_n_rollercoaster.csv")
toy_story_mania_df = pd.read_csv("./Disney Resources/toy_story_mania (1).csv")


# Database of select rides and notes of which Theme Park they belong to 
# Each of these rides are individual tables

print(seven_dwarfs_df) # Magic kingdom 
print(alien_saucers_df) # Hollywood Studios
print(entities_df) 
print(metadata_df)
print(pirates_of_caribbean_df) # Magic Kingdom
print(rock_n_rollercoaster_df) # Hollywood Studios
print(toy_story_mania_df) # Hollywood Studios

````

### Magic Kingdom


````python
# Data Cleanup, this includes checking for any NaN and negative values, and replacing them with the value of 0
# Will need to repeat this process for each individual ride dataframe

seven_dwarfs_df = seven_dwarfs_df.fillna(0)

#seven_dwarfs_df["SACTMIN"].value_counts()
#seven_dwarfs_df["SPOSTMIN"].value_counts()

seven_dwarfs_df["SPOSTMIN"] = seven_dwarfs_df["SPOSTMIN"].replace(-999, 0)
seven_dwarfs_df["SACTMIN"] = seven_dwarfs_df["SACTMIN"].replace(-92918.0, 0)

seven_dwarfs_df["SPOSTMIN"].value_counts()
list(seven_dwarfs_df["SACTMIN"].value_counts().index)
seven_dwarfs_df["SACTMIN"].value_counts()


# Rename column titles 
seven_dwarfs_df = seven_dwarfs_df.rename(columns={"SPOSTMIN": "Posted_Wait_Time", "SACTMIN": "Actual_Wait_Time"})
seven_dwarfs_df

# Store dataframe to csv as backup
#seven_dwarfs_df.to_csv('seven_dwarfs_backup.csv')
````

````python
pirates_of_caribbean_df

pirates_of_caribbean_df["SPOSTMIN"].value_counts()
pirates_of_caribbean_df = pirates_of_caribbean_df.fillna(0)

list(pirates_of_caribbean_df["SACTMIN"].value_counts().index)

pirates_of_caribbean_df["SACTMIN"] = pirates_of_caribbean_df["SACTMIN"].replace(-7, 0)
pirates_of_caribbean_df["SPOSTMIN"] = pirates_of_caribbean_df["SPOSTMIN"].replace(-999.0, 0)

list(pirates_of_caribbean_df["SPOSTMIN"].value_counts().index)

pirates_of_caribbean_df = pirates_of_caribbean_df.rename(columns={"SPOSTMIN": "Posted_Wait_Time", "SACTMIN": "Actual_Wait_Time"})
pirates_of_caribbean_df

````
